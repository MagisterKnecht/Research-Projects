```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #I love the style!
os.chdir('C:\\Users\\baxte\\Documents')
pd.set_option('precision',10)
```

```python
ChinaData = pd.read_csv('ChinaData.csv')
```

```python
fig = plt.figure(figsize=(20,10))

plt.plot(ChinaData['Year'], ChinaData['NEO'], color='crimson', marker='o')
plt.xlabel('Year', fontdict=None, labelpad=None, size='20')
plt.ylabel('Millions of USD', fontdict=None, labelpad=None, size='20')
plt.title('Chinese Net Errors and Omissions, 1984-2016', fontdict=None, size='25')
#This plot shows the increase in Net Errors and Omissions. A positive increase shows people moving their money from the country.
```

```python
plt.figure(figsize=(20,10))
plt.plot(ChinaData['Year'], ChinaData['NEObyTotalReserves'], marker='o', color='mediumvioletred')
plt.xlabel('Year', fontdict=None, labelpad=None, size='20')
plt.ylabel('Ratio of NEO to Total Reserves minus Gold', fontdict=None, labelpad=None, size='20')
plt.title('Ratio of Chinese NEO to Total Reserves minus Gold, 1984-2016', fontdict=None, size='25')
#This plot shows the increase in Net Errors and Omissions. A positive increase shows people moving their money from the country.
```
